```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load citra grayscale
image = cv2.imread('kucing.png', 0)
```

```python
# Buat kernel Roberts
kernel_45 = np.array([[1, 0],
                      [0, -1]], dtype=np.float32)

kernel_135 = np.array([[0, 1],
                       [-1, 0]], dtype=np.float32)
```

```python
# Terapkan filter Roberts pada citra
filtered_45 = cv2.filter2D(image, -1, kernel_45)
filtered_135 = cv2.filter2D(image, -1, kernel_135)
```

```python
# Kombinasikan hasil dari kedua filter
result = np.sqrt(filtered_45**2 + filtered_135**2)
```

```python
# Konversi hasil ke dalam skala 0-255
result = np.uint8(result)
```

```python
#Asli
plt.subplot(2,2,1), plt.imshow(image, cmap='gray')
plt.title('Citra Asli'), plt.xticks([]),plt.yticks([])

#Roberts
plt.subplot(2,2,2), plt.imshow(result, cmap=plt.get_cmap('gray'))
plt.title('Roberts')
plt.show()
```

```python
plt.imshow(image, cmap=plt.get_cmap('gray'))
plt.show()
```

```python
plt.imshow(result, cmap=plt.get_cmap('gray'))
plt.show()
```

```python
img = cv2.imread('rumah.jpeg', 0)
```

```python
filtered_452 = cv2.filter2D(img, -1, kernel_45)
filtered_1352 = cv2.filter2D(img, -1, kernel_135)
```

```python
# Kombinasikan hasil dari kedua filter
resultd = np.sqrt(filtered_452**2 + filtered_1352**2)
resultd = np.uint8(resultd)
```

```python
#Asli
plt.subplot(2,2,1), plt.imshow(img, cmap='gray')
plt.title('Citra Asli'), plt.xticks([]),plt.yticks([])

#Roberts
plt.subplot(2,2,2), plt.imshow(resultd, cmap=plt.get_cmap('gray'))
plt.title('Roberts')
plt.show()
```
