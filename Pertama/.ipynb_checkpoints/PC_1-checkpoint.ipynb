```python
import matplotlib, cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = cv2.imread('skyy.jpg')
im = mpimg.imread('skyy.jpg')
```

```python
#convert image to RGB color for matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#show image with matplotlib
plt.imshow(img)
```

```python
#convert image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

#grayscale image represented as a 2-d array
print(gray_img)
```

```python
#have to convert grayscale back to RGB for plt.imshow()
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_GRAY2RGB))
```

```python
#Convert a grayscale image to black and white using binary thresholding 
(thresh, BnWimg) = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(BnWimg)
```
