```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load grayscale image
img_gray = cv2.imread('fotbar.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
# Calculate histogram
hist_gray, bins_gray = np.histogram(img_gray.flatten(), 256, [0,256])
```

```python
# Plot histogram
plt.hist(img_gray.flatten(), 256, [0,256], color='gray')
plt.xlim([0,256])
plt.xlabel('Gray Level')
plt.ylabel('Pixel Count')
plt.title('Gray Scale Histogram')
plt.show()
```

```python
# Histogram Equalization
img_eq = cv2.equalizeHist(img_gray)
```

```python
# Calculate histogram after equalization
hist_eq, bins_eq = np.histogram(img_eq.flatten(), 256, [0,256])
```

```python
# Plot histogram after equalization
plt.hist(img_eq.flatten(), 256, [0,256], color='gray')
plt.xlim([0,256])
plt.xlabel('Gray Level')
plt.ylabel('Pixel Count')
plt.title('Histogram Equalization')
plt.show()
```
